above there is issue with Ship Mode as it has values of 'Not Avaiable' and 'unknown'

So,we will read the above csv as if by default we are treating 'Not Available' and 'unknown' as nan

```python
df1=pd.read_csv('orders_data.csv',na_values=['Not Available','unknown'])
df1['Ship Mode'].unique()
```

### Calculating the Selling Price  and Profit

```python
df1['Selling Price']=df1['List Price']-(df1['List Price']*df1['Discount Percent']/100)
df1['Selling Price']
df1.head()
```

```python
df1['Profit']=df1['Selling Price']-df1['cost price']
```

```python
df1.rename(columns={'Profit':'Unit_Profit','Selling Price':'Unit_Selling_Price'},inplace=True)
df1.head()
```

```python
df1.dtypes
```

### Changing Datatype of Order Date from object to datetime

```python
df1['Order Date']=pd.to_datetime(df1['Order Date'],format='%d-%m-%Y')
```

```python
df1.dtypes
```

```python
df1.head()
```

### Dropping few **used or unnecessary** columns

```python
df1.drop(columns=['cost price','List Price','Discount Percent'],inplace=True)
df1.head(2)
```

Calculating **Total Profit** in each order

```python
df1['Total Profit']=df1['Unit_Profit']*df1['Quantity']
df1.head(2)
```

### Q15. Find month over month growth comparison for 2022 and 2023 sales eg : jan 2022 vs jan 2023

```python
df1.head()
```

```python
a=df1.groupby(['Month', 'Year']).sum('Total_Sale')[['Total_Sale']]
a
```

```python
pivot_table = a.pivot_table(index='Month', columns='Year', values='Total_Sale', aggfunc='sum')
pivot_table
```

```python
pivot_table['YoY_growth'] = pivot_table[2023] - pivot_table[2022]
pivot_table
```

### Q17. Which sub category had highest growth by profit in 2023 compare to 2022

```python
m = df1.groupby(['Sub Category', 'Year']).sum(['Total Profit'])['Total Profit'].reset_index()
m
```

```python
n = m.pivot_table(index = 'Sub Category', columns='Year', values = 'Total Profit').reset_index()
n
```

```python
n['diff'] = n[2023] - n[2022]

n.sort_values(by = ['diff'], ascending=False)
```
